```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("05-Seaborn/dm_office_sales.csv")
```

```python
df.head()
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
sns.set(style='darkgrid')
sns.displot(data=df,x='salary',kde=True)
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=['Age'])
```

```python
sample_ages.head()
```

```python
sns.rugplot(x='Age',data=sample_ages)
```

```python
sns.displot(data=sample_ages,x='Age',rug=True,kde=True)
```

```python
sns.kdeplot(data=sample_ages,x='Age',clip=[0,100],shade=True)
```

# Categorical plots

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df,x='level of education',hue='division',palette='Set2')
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df,y='salary',x='level of education',estimator=np.mean,ci='sd',hue='division')
plt.legend(bbox_to_anchor=(1.05,1))
```
