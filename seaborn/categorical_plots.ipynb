```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv("05-Seaborn/StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,y='reading score',x='parental level of education',hue='test preparation course',palette='Set2')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.violinplot(data=df,x='reading score',y='parental level of education',bw=0.1)
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.swarmplot(data=df,x='math score',size=2,y='gender',
             hue='test preparation course')
plt.legend(bbox_to_anchor=(1.1,0.5))
```
