```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv("05-Seaborn/StudentsPerformance.csv")
```

```python
df.head()
```

```python
sns.catplot(data=df,y='math score',x='gender',kind='box',col='lunch',row='test preparation course')
```

```python
sns.pairplot(df)
```

```python
g = sns.PairGrid(df)

g.map_upper(sns.scatterplot)

g.map_lower(sns.kdeplot)

g.map_diag(sns.kdeplot)
```
