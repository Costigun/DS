```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
```

```python
b = a ** 4
```

```python
x = np.arange(0,10)
```

```python
y = 2 * x
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[0][0].set_xlabel('X 0,0')

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)

axes[1][1].plot(a,b)
axes[1][1].set_title('My Title 1,1')
axes[1][1].set_xlim(2,6)


fig.suptitle('Figure level',fontsize=16)
plt.tight_layout()
```
