```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# ax.plot(x,x,color='#4a8c5b',label='x vs x',lw=10,ls=':') #RGB HEX CODE

ax.plot(x,x+1,color='purple',label='x vs x + 1',linewidth=2,marker='s',ls='--',ms=15,
       markerfacecolor='red',markeredgewidth=4,markeredgecolor='orange')


plt.legend()
```
