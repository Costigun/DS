```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
```

```python
b = a ** 4
```

```python
x = np.arange(0,10)
```

```python
y = 2 * x
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')


axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(a,b)
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

```

```python
fig = plt.figure(figsize=(2,2),dpi=200)

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

fig.savefig('newfigure.png',bbox_inches='tight')
```
