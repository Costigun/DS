```python
import numpy as np
import pandas as pd
```

```python
df_2018 = pd.read_excel('документы приобретения товаров за 2018 год.xls')
df_2019 = pd.read_excel('документ приобретения товаров за 2019 год.xls')
df_2020 = pd.read_excel('документ приобретения товара за 2020 год.xls')
```

```python
df_2018 = df_2018.drop('Unnamed: 0',axis=1)
df_2019 = df_2019.drop('Unnamed: 0',axis=1)
df_2020 = df_2020.drop('Unnamed: 0',axis=1)
```

```python
sum_2018 = df_2018['Сумма'].sum()
sum_2019 = df_2019['Сумма'].sum()
sum_2020 = df_2020['Сумма'].sum()
print("Сумма операцийй за 2018 год = {} \n Сумма операцийй за 2019 год = {} \n Сумма операцийй за 2020 год = {}".format(sum_2018,sum_2019,sum_2020))
```

```python
df_2019_rs = df_2019.groupby('Контрагент').sum()
result_2019 = df_2019_rs.nlargest(1,'Сумма')
result_2019['Год'] = 2019
```

```python
df_2018_rs = df_2018.groupby('Контрагент').sum()
result_2018 = df_2018_rs.nlargest(1,'Сумма')
result_2018['Год'] = 2018
```

```python
df_2020_rs = df_2020.groupby('Контрагент').sum()
result_2020 = df_2020_rs.nlargest(1,'Сумма')
result_2020['Год'] = 2020
```

```python
df_2018_rs = df_2018.groupby('Контрагент').sum()
df_2018_rs.agg(['sum','mean','max'])
```

```python
frames = [result_2018,result_2019,result_2020]
```

```python
df = pd.concat(frames)
```
